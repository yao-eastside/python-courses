# randn vs rand vs random

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.random.randn(10000)
x1[:5]
```

```python
plt.plot(x1[:10000].cumsum())
```

```python
pd.DataFrame(x1).plot.hist()
```

```python
x2 = np.random.rand(1000)
x2[:5]
```

```python
pd.DataFrame(x2).plot.hist()
```

```python
x3 = np.random.random(1000)
x3[:5]
```

```python
pd.DataFrame(x3).plot.hist()
```
